```python
# get h mart business ids from dataset_business.json
df_h = df_b[(df_b['name'].str.lower().str.contains('h mart|hmart|h-mart'))&(~df_b['business_id'].isin(['WUZsRFKkL2YKdCp_SHNzqQ','U_y_qiiU-aIguP-RWWaCXQ','ofzlAnK5txsaXWpfSwXhug']))]
hmart_ids = df_h['business_id'].to_list()

# create a dictionary that can save reviews by stores
hmart_reviews = {}
for id in hmart_ids:
    hmart_reviews[id] = None
```

```python
hmart_reviews
```

```python
# get reviews from dataset_reviews.json using above business ids
# save as pandas dataframe in the dictionary
with open('yelp_academic_dataset_review.json', encoding='utf-8') as f:
    for line in f:
        for id in hmart_reviews.keys():
            if id in line:
                hmart_reviews[id].append(line)
```

```python
# check how many reviews found by store and total
total = 0

for id in hmart_reviews.keys():
    print(f'{id}: {len(hmart_reviews[id])}')
    total += len(hmart_reviews[id])

print(f'\n total: {total}')
```
